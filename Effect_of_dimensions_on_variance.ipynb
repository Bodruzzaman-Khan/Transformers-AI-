```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Generate 1000 pairs of 50-dimensional random vectors
num_vectors = 1000
vector_dim = 150

# Random vectors
vectors_a = np.random.randn(num_vectors, vector_dim)
vectors_b = np.random.randn(num_vectors, vector_dim)

# Compute the dot products of each pair
dot_products = np.einsum('ij,ij->i', vectors_a, vectors_b)

# Plot the histogram of dot products
plt.figure(figsize=(8, 6))
plt.hist(dot_products, bins=30, color='purple')
plt.title('Dot Product Distribution (1000 pairs of 50-dimensional vectors)')
plt.xlabel('Dot Product Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

# Define the different dimensions for the vectors
vector_dims = [2, 10, 100, 1000]  # Updated vector dimensions
num_vectors = 1000  # Number of vector pairs

# Create a color map for the histograms
colors = ['purple', 'green', 'blue', 'orange']
labels = [f'{dim}-Dimensional' for dim in vector_dims]

# Set up the plot
plt.figure(figsize=(10, 6))

# Loop through each dimension, compute dot products, and plot histogram
for idx, vector_dim in enumerate(vector_dims):
    # Generate random vectors for the current dimension
    vectors_a = np.random.randn(num_vectors, vector_dim)
    vectors_b = np.random.randn(num_vectors, vector_dim)

    # Compute the dot products
    dot_products = np.einsum('ij,ij->i', vectors_a, vectors_b)

    # Plot the histogram for the current dimension with outline
    plt.hist(dot_products, bins=20, color=colors[idx], histtype='step', linewidth=2, label=f'{vector_dim}-Dimensional')

# Add plot labels and legend
plt.xlabel('Dot Product Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# Show the combined plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Set the font properties
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'DejaVu Serif']  # Add a fallback font

# Define the different dimensions for the vectors
vector_dims = [2, 5, 10, 20, 50, 100, 200, 500, 800, 1000, 1500, 2000]  # Expanded vector dimensions
num_vectors = 2000  # Number of vector pairs
colors = ['purple', 'green', 'blue', 'orange', 'red', 'brown', 'yellow', 'black', 'teal', 'navy', 'maroon', 'olive']
labels = [f'{dim}-Dimensional' for dim in vector_dims]

# Set up the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Title and labels
ax.set_xlabel('Dot Product Value')
ax.set_ylabel('Frequency')
ax.grid(False)

# Function to generate dot products and return values
def generate_dot_products(vector_dim):
    vectors_a = np.random.randn(num_vectors, vector_dim)
    vectors_b = np.random.randn(num_vectors, vector_dim)
    dot_products = np.einsum('ij,ij->i', vectors_a, vectors_b)
    return dot_products

# Update function for animation
def update(frame):
    # Clear previous histograms
    ax.clear()

    # Redraw the title and labels (since they are cleared too)
    # ax.set_title('Comparison of Dot Product Distributions Across Dimensions', fontsize=16)
    ax.set_xlabel('Dot Product Distribution\n\n\n $\it{Bodruzzaman\ Khan}$', fontsize=14)
    ax.set_ylabel('Frequency', fontsize=14)
    ax.grid(False)

    # Plot histograms up to the current frame
    for i in range(frame + 1):
        dot_products = generate_dot_products(vector_dims[i])
        ax.hist(dot_products, bins=20, color=colors[i], histtype='step', linewidth=3, label=labels[i])

    # Set x-axis limits to keep the window centered around 0
    ax.set_xlim(-170, 170)  # Adjust this range as needed

    # Add the legend each frame
    ax.legend()

# Create the animation with slower speed (interval controls speed)
ani = FuncAnimation(fig, update, frames=len(vector_dims), repeat=False, interval=800)  # Increased interval to 800ms

# Save the animation locally as a .html file
ani.save('dot_product_distribution_bodruzzaman.mp4', writer='ffmpeg',
        #  fps=10
         )


# Display the animation in Colab using HTML
HTML(ani.to_jshtml())
```
